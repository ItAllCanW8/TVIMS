```python
import random
import math
import matplotlib.pylab as plt
```

# Task 1

```python
def foo(n, t_quantile):
    print(
        "+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=++=+=+=+=+=+=+=+=+=+=+==+=+=+=+=+=++=+=+=+=+=+=+==+=+=+=+=+=+=+=+=+=+=+=+=+=+")
    print(f"n = {n}")

    x_arr = []
    y_arr = []

    values = [0, math.pi / 6, math.pi / 4, math.pi / 3, math.pi / 2]

    l_trust_limit = []
    h_trust_limit = []
    trust_intervals = []

    t_l_trust_limit = []
    t_h_trust_limit = []
    t_trust_intervals = []

    sign_level = [0.9, 0.95, 0.99, 0.999]

    M = 0
    D = 0

    x_arr += [random.choice(values) for i in range(n)]
    y_arr += [round(math.sin(x), 2) for x in x_arr]
    y_arr.sort()

    print(f"Sample: {y_arr}")

    for y in y_arr:
        M += y

    M /= n
    M = round(M, 3)

    print(f"Estimate of mathematical expectation: {M}")

    for y in y_arr:
        D += (y - M) ** 2

    D /= (n - 1)
    D = round(D, 3)

    print(f"Unbiased dispersion estimate: {D}")

    tolerance = 0

    for lvl in t_quantile:
        tolerance = round(lvl * ((D / n) ** 0.5), 3)
        l_trust_limit += [M - tolerance]
        h_trust_limit += [M + tolerance]
        trust_intervals += [2 * tolerance]

    i = 0

    for lvl in sign_level:
        print(f"For sign. lvl = {lvl}: {l_trust_limit[i]} <= m*(y) <= {h_trust_limit[i]}")
        i += 1

    plt.xlabel("significance lvl")
    plt.ylabel("trust interval value")
    plt.plot(sign_level, trust_intervals)
    plt.show()

    # Y c [0;1], assume uniform distribution from Zadanie 1, D = (b - a)^2 / 12
    theor_D = (1 - 0) ** 2 / 12

    tolerance = 0

    for lvl in t_quantile:
        tolerance = round(lvl * ((theor_D / n) ** 0.5), 3)
        t_l_trust_limit += [M - tolerance]
        t_h_trust_limit += [M + tolerance]
        t_trust_intervals += [2 * tolerance]

    i = 0

    for lvl in sign_level:
        print(
            f"For sign. lvl = {lvl} and theoretical D = {theor_D}: {t_l_trust_limit[i]} <= m*(y) <= {t_h_trust_limit[i]}")
        i += 1

    plt.xlabel("significance lvl")
    plt.ylabel("theor. trust interval value")
    plt.plot(sign_level, t_trust_intervals)
    plt.show()

    plt.xlabel("significance lvl")
    plt.plot(sign_level, trust_intervals)
    plt.plot(sign_level, t_trust_intervals)
    plt.show()

    print(
        "+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=++=+=+=+=+=+=+=+=+=+=+==+=+=+=+=+=++=+=+=+=+=+=+==+=+=+=+=+=+=+=+=+=+=+=+=+=+")

    return trust_intervals[0]


# for significance value = 0.9
trust_intervals = []
n = [20, 30, 50, 70, 100, 150]

t_quantiles = [[1.72913281152, 2.09302405441, 2.86093460645, 3.8834058494],
               [1.69912702653, 2.04522964213, 2.75638590367, 3.6594050194],
               [1.67722419534, 2.01063475469, 2.68220402695, 3.5050679704],
               [1.667572280661, 1.99546893091, 2.65008129281, 3.4394354769],
               [1.660551217044, 1.98446745442, 2.62693109481, 3.3925881214],
               [1.6552145061759864, 1.9761224, 2.60945633117, 3.3574777920]]

trust_intervals += [foo(n[0], t_quantiles[0])]
trust_intervals += [foo(n[1], t_quantiles[1])]
trust_intervals += [foo(n[2], t_quantiles[2])]
trust_intervals += [foo(n[3], t_quantiles[3])]
trust_intervals += [foo(n[4], t_quantiles[4])]
trust_intervals += [foo(n[5], t_quantiles[5])]

plt.xlabel("n")
plt.ylabel("trust interval value")
plt.plot(n, trust_intervals)
plt.show()
```

# Task 2

```python
def foo(n, l_quantile, h_quantile, l_quantile_2, h_quantile_2):
    print(f"n = {n}")
    x_arr = []
    y_arr = []

    values = [0, math.pi / 6, math.pi / 4, math.pi / 3, math.pi / 2]

    l_trust_limit = [0, 0, 0]
    h_trust_limit = [0, 0, 0]
    trust_intervals = [0, 0, 0]

    t_l_trust_limit = [0, 0, 0]
    t_h_trust_limit = [0, 0, 0]
    t_trust_intervals = [0, 0, 0]

    sign_level = [0.9, 0.95, 0.99]

    n = 20

    M = 0
    D = 0

    x_arr += [random.choice(values) for i in range(n)]
    y_arr += [round(math.sin(x), 2) for x in x_arr]
    y_arr.sort()

    print(f"Sample: {y_arr}")

    for y in y_arr:
        M += y

    M /= n
    M = round(M, 3)

    print(f"Estimate of mathematical expectation: {M}")

    for y in y_arr:
        D += (y - M) ** 2

    D /= (n - 1)
    D = round(D, 3)

    print(f"Unbiased dispersion estimate: {D}")

    i = 0

    for i in range(3):
        l_trust_limit[i] = (n - 1) * D / l_quantile[i]
        h_trust_limit[i] = (n - 1) * D / h_quantile[i]
        trust_intervals[i] = h_trust_limit[i] - l_trust_limit[i]

    i = 0

    for lvl in sign_level:
        print(f"For sign. lvl = {lvl}: {l_trust_limit[i]} <= D*(y) <= {h_trust_limit[i]}")
        i += 1

    plt.xlabel("significance lvl")
    plt.ylabel("trust interval value")
    plt.plot(sign_level, trust_intervals)
    plt.show()

    # Y c [0;1], assume uniform distribution from Zadanie 1, M = (a + b) / 2
    theor_M = (0 + 1) / 2
    theor_D = 0

    for y in y_arr:
        theor_D += (y - theor_M) ** 2

    theor_D /= (n - 1)
    theor_D = round(theor_D, 3)

    i = 0

    for i in range(3):
        t_l_trust_limit[i] = n * theor_D / l_quantile_2[i]
        t_h_trust_limit[i] = n * theor_D / h_quantile_2[i]
        t_trust_intervals[i] = t_h_trust_limit[i] - t_l_trust_limit[i]

    i = 0

    for lvl in sign_level:
        print(
            f"For sign. lvl = {lvl} and theoretical M = {theor_M}: {t_l_trust_limit[i]} <= D*(y) <= {t_h_trust_limit[i]}")
        i += 1

    plt.xlabel("significance lvl")
    plt.ylabel("theor. trust interval value")
    plt.plot(sign_level, t_trust_intervals)
    plt.show()

    plt.xlabel("significance lvl")
    plt.plot(sign_level, trust_intervals)
    plt.plot(sign_level, t_trust_intervals)
    plt.show()

    print(
        "+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=+=++=+=+=+=+=+=+=+=+=+=+==+=+=+=+=+=++=+=+=+=+=+=+==+=+=+=+=+=+=+=+=+=+=+=+=+=+")

    return trust_intervals[0]


# for significance level = 0.9
trust_intervals = []
n = [20, 30, 50, 70, 100, 150]

l_quantiles = [[31.41, 34.17, 40], [42.55, 45.72, 52.33]]
h_quantiles = [[10.85, 9.59, 7.43], [17.70, 16.04, 13.12]]

l_quantiles_2 = [[32.67, 35.47, 41.40], [43.77, 46.97, 53.67]]
h_quantiles_2 = [[11.59, 10.28, 8.03], [18.49, 16.79, 13.78]]

trust_intervals += [foo(n[0], l_quantiles[0], h_quantiles[0], l_quantiles_2[0], h_quantiles_2[0])]
trust_intervals += [foo(n[1], l_quantiles[1], h_quantiles[1], l_quantiles_2[1], h_quantiles_2[1])]
trust_intervals += [foo(n[2], l_quantiles[1], h_quantiles[1], l_quantiles_2[1], h_quantiles_2[1])]
trust_intervals += [foo(n[3], l_quantiles[1], h_quantiles[1], l_quantiles_2[1], h_quantiles_2[1])]
trust_intervals += [foo(n[4], l_quantiles[1], h_quantiles[1], l_quantiles_2[1], h_quantiles_2[1])]
trust_intervals += [foo(n[5], l_quantiles[1], h_quantiles[1], l_quantiles_2[1], h_quantiles_2[1])]

plt.xlabel("n")
plt.ylabel("trust interval value")
plt.plot(n, trust_intervals)
plt.show()
```
